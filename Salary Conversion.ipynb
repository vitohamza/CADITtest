```python
import requests
import pandas as pd
import json
```

```python
userjs = requests.get('http://jsonplaceholder.typicode.com/users')
user = userjs.json()
```

```python
salarydata=[]
salary = pd.read_json('D:\Vito\Work Prep\CADIT\salary_data.json')
for i in range(10):
    userid = salary['array'][i]['id']
    usersalary = salary['array'][i]['salaryInIDR']
    salarydata.append([userid, usersalary])
salarytable = pd.DataFrame(salarydata)
salarytable.columns = ['id','salaryIDR']
```

```python
userdata = []
for i in range(10):
    id = user[i]['id']
    name = user[i]['name']
    uname = user[i]['username']
    email = user[i]['email']
    address = str(user[i]['address']['street'])+", "+str(user[i]['address']['suite'])+", "+str(user[i]['address']['city'])+", "+str(user[i]['address']['zipcode'])+", "+str(user[i]['address']['geo'])
    phone =  user[i]['phone']
    userdata.append([id, name, uname, email, address, phone])
usertable = pd.DataFrame(userdata)
usertable.columns = ['id','name', 'username', 'email', 'address', 'phone']
```

```python
userslry = pd.merge(usertable, salarytable, on='id')
userslry
```

```python
convert = requests.get('https://free.currconv.com/api/v7/convert?apiKey=f3b2235e37e079dd4905&q=IDR_USD&compact=y').json()
convIDR = convert['IDR_USD']['val']
```

```python
usd = pd.DataFrame(userslry['salaryIDR']*convIDR)
usd.columns = ['salaryUSD']
output = pd.concat([userslry,usd], axis = 1)
#usertable.columns = ['id','USD']
#userslry = pd.merge(userslry, usd, on='id')
```
